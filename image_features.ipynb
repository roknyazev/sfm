```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('data/01.jpg',cv.IMREAD_GRAYSCALE)
plt.imshow(img)
```

### scale

```python
print(img.shape)
scale_percent = 35 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
img = cv.resize(img, dim, interpolation = cv.INTER_AREA)
img.shape
```

### FAST

```python
fast = cv.FastFeatureDetector_create()
kp = fast.detect(img,None)
img2 = cv.drawKeypoints(img, kp, None, color=(255,0,0))
plt.imshow(img2)
```

### ORB

```python
orb = cv.ORB_create()
kp = orb.detect(img,None)
kp, des = orb.compute(img, kp)
img3 = cv.drawKeypoints(img, kp, None, color=(255,0,0), flags=0)
plt.imshow(img3)
```

### SIFT

```python
sift = cv.SIFT_create()
kp = sift.detect(img,None)
img4=cv.drawKeypoints(img, kp, None, color=(255,0,0), flags=0)
plt.imshow(img4)
```

### BRIEF

```python
star = cv.xfeatures2d.StarDetector_create()
brief = cv.xfeatures2d.BriefDescriptorExtractor_create()
kp = star.detect(img,None)
kp, _ = brief.compute(img, kp)
img5 = cv.drawKeypoints(img, kp, None, color=(255,0,0), flags=0)
plt.imshow(img5)
```
